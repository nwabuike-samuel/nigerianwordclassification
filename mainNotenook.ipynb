```python
# Join all words into a single string and convert to lowercase
all_letters = ''.join(words).lower()

# Count frequency of each character
frequency = Counter(all_letters)

# Filter to include only alphabet characters
frequency = {char: count for char, count in frequency.items() if char.isalpha()}

print(frequency)

# Sort characters by frequency in descending order
sorted_chars = sorted(frequency, key=frequency.get, reverse=True)

print(sorted_chars)

# Map each character to an index based on frequency
char_to_index = {char: idx for idx, char in enumerate(sorted_chars, start=1)}

print(char_to_index)
```

```python
model = Sequential([
    # He Normal initializer for weights, zero initializer for biases
    Dense(16, input_dim=12, activation='relu', kernel_initializer=HeNormal(), bias_initializer='zeros'),
    # Dropout(0.5),
    # Another dense layer with different initializers
    Dense(32, activation='relu', kernel_initializer=HeNormal(), bias_initializer='ones'),
    Dense(64, activation='relu', kernel_initializer=HeNormal(), bias_initializer='ones'),
    Dense(32, activation='relu', kernel_initializer=HeNormal(), bias_initializer='ones'),
    Dense(16, activation='relu', kernel_initializer=HeNormal(), bias_initializer='ones'),
    # Dense(8, activation='relu', kernel_initializer=HeNormal(), bias_initializer='ones'),
    # Dropout(0.5),
    # Output layer
    Dense(1, activation='sigmoid', kernel_initializer='glorot_normal', bias_initializer=Constant(value=0.1))  # small constant for bias
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```
